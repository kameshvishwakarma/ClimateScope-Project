   Monthly Average Temperature (df['last_updated'].dt.year.unique())

```python
monthlyTemperature = df.groupby('month')['temperature_celsius'].mean()
monthlyTemperature
```

```python
plt.figure()
monthlyTemperature.plot(kind='line')
plt.title("Average Monthly Temperature")
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")
plt.show()
```

 Monthly Average Rainfall 

```python
monthlyRain = df.groupby('month')['precip_mm'].mean()
monthlyRain
```

```python
plt.figure()
monthlyRain.plot(kind='line')
plt.title("Average Monthly Rainfall")
plt.xlabel("Month")
plt.ylabel("Rainfall (mm)")
plt.show()
```

Extreame Weather Event

```python
extremeHotTemp = df[df['temperature_celsius'] > 45]
```

```python
#extremeHotTemp
extremeHotTemp[['country','temperature_celsius']].head(10) 
```

```python
extremeHotTemp['country'].value_counts()
```

```python
extreme_count = extremeHotTemp['country'].value_counts()
plt.figure()
extreme_count.plot(kind='bar')
plt.title("Extreme Heat Temperature >45°C by Country")
plt.xlabel("Country")
plt.ylabel("count")
plt.show()
```
